```python
from sklearn.externals import joblib
from sklearn.feature_selection import SelectFromModel
import numpy as np
```

```python
CL_SOC = joblib.load("../trained_models/CL_SOC.pkl")
```

```python
cls = CL_SOC
decision = cls.classifier.steps[-1][-1]
```

```python
feature_names = np.asarray(cls.feature_extractor("Bill")[-1])
```

```python
importance = -1*np.linalg.norm(decision.coef_,axis=0)
feature_ranks = feature_names[np.argsort(importance)].flatten()
print("\n".join(feature_ranks))
```
