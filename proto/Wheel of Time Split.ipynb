```python
from book import *
import os.path
import os
import subprocess
import shlex
import re
```

```python
def clean_book(input_filepath, output_filepath):
    input_ext = os.path.splitext(input_filepath)[1]
    output_ext = os.path.splitext(output_filepath)[1]
    assert (output_ext == ".epub")

    print()
    print("-"*30)
    print(input_filepath)
    print(output_filepath)
    print()
    
    cmd = ['ebook-convert',
           shlex.quote(input_filepath),
           shlex.quote(output_filepath),
           '--enable-heuristics',
           "--chapter",
           """ "//*[((name()='h1' or name()='h2'  or name()='span') and re:test(., '\s*((chapter|CHAPTER|Chapter)\s+)|((prolog|prologue|epilogue|PROLOG|PROLOGUE|EPILOGUE|Prologue)(\s+|$))', 'i')) or @class = 'chapter' or @class = 'scenebreak']" """   
          ]

    cmd =  " ".join(cmd)
    subprocess.run(cmd, shell=True)
        
```

```python
def ischapter_text(beginning):
    # just look for the word chapter occurring in first 20 characters
    # We'll skip the prologs etc
    return re.match(r".*chapter", beginning, flags=re.IGNORECASE + re.DOTALL)
```

```python
ann_WOT = []
for bookname in books_to_use:
    texts = load_chapters(load_book(book2path(bookname)))[0]
    
    
    for (ii, text) in enumerate(texts):
        beginning = text[:20]
        
        if ischapter_text(beginning):
            chapter_num = re.findall(r"\d+", beginning, flags=re.I)[0]
            
            chapter2name = book2chapter2name[bookname]                    
            character = chapter2name[chapter_num]
            
            if character == "Narrator":
                print("skipping due to being for Narrator", ii, bookname, " (no prob if prolog etc)")
                print(text.strip()[:512])
                continue
            
            if len(text)<256:
                if ii+1<len(texts) and len(texts[ii+1])>=256 and not ischapter_text(texts[ii+1][:20]):
                    # Try using the next item, maybe chapters got split up.
                    text = text + texts[ii+1]
                    print("repairing ", ii, bookname,character, " to")
                    print(text.strip()[:512])
                    print("++")
                else:
                    print("skipping due to shortness: ", ii, bookname,character, "(if this ocurs late but chapter does earlier/after, worries probably a contents page)")
                    print(text.strip())
                    print("++")
                    continue
            else:
                print(ii, bookname, chapter_num)
            ann_WOT.append({'character':character, 'text':text})
        
        

import json
with open("../flat_data/WOT" + ".json",'w') as fh:
    json.dump(ann_WOT, fh)
```
